### 5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generating a synthetic dataset with correlated features
np.random.seed(42)
data_size = 100
feature_1 = np.random.rand(data_size) * 100
feature_2 = 2.5 * feature_1 + np.random.randn(data_size) * 10
feature_3 = 0.5 * feature_1 - np.random.randn(data_size) * 5
feature_4 = 1.2 * feature_2 + np.random.randn(data_size) * 2

# Creating a dataset by combining the features
data = np.array([feature_1, feature_2, feature_3, feature_4]).T

# Creating a correlation matrix
correlation_matrix = np.corrcoef(data.T)

# Visualizing the correlation matrix using a heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', xticklabels=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'], yticklabels=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'])

# Add title
plt.title('Correlation Matrix Heatmap')

# Show the heatmap
plt.show()
```

### 1. Using the given dataset, to generate a 3D scatter plot to visualize the distribution of data points in a threedimensional space

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Setting the seed and generating the data
np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

# Creating the DataFrame
df = pd.DataFrame(data)

# Creating a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['X'], df['Y'], df['Z'], c='b', marker='o')

# Setting labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

### 2. Using the Student Grades, create a violin plot to display the distribution of scores across different grade categories


```python
np.random.seed(15)
sales_data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df_sales = pd.DataFrame(sales_data)

# Creating a violin plot to visualize the distribution of sales across different months
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.violinplot(x='Month', y='Sales', data=df_sales)
plt.title('Violin Plot of Sales Distribution Across Different Months')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.show()
```

### 3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days

```python
# Creating a pivot table for the heatmap based on Month and Day
sales_pivot = df_sales.pivot_table(index='Month', columns='Day', values='Sales', aggfunc='mean')

# Creating a heatmap to visualize the variation in sales across different months and days
plt.figure(figsize=(12, 6))
sns.heatmap(sales_pivot, cmap="YlGnBu", annot=True, fmt=".0f", linewidths=.5)

plt.title('Heatmap of Sales Across Different Months and Days')
plt.xlabel('Day')
plt.ylabel('Month')

plt.show()
```

### 4. Using the given x and y data, generate a 3D surface plot to visualize the function

```python
# Generating the 3D surface plot
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Creating a 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Setting labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.title('3D Surface Plot of sin(sqrt(x^2 + y^2))')
plt.show()
```


### 5. Using the given dataset, create a bubble chart to represent each country's population (y-axis), GDP (xaxis), and bubble size proportional to the population


```python
import matplotlib.pyplot as plt

# Generating data for the bubble chart
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df_countries = pd.DataFrame(data)

# Creating a bubble chart
plt.figure(figsize=(8, 6))
plt.scatter(df_countries['GDP'], df_countries['Population'], s=df_countries['Population']*10, alpha=0.5)

# Adding labels and title
for i, country in enumerate(df_countries['Country']):
    plt.text(df_countries['GDP'][i], df_countries['Population'][i], country)

plt.xlabel('GDP (in billions)')
plt.ylabel('Population (in millions)')
plt.title('Bubble Chart of Population vs GDP for Countries')

plt.show()
```

# **BOKEH ASSIGNMENT:**

### 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x

```python
import numpy as np
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource

# Prepare the data
x = np.linspace(0, 10, 500)
y = np.sin(x)

# Create a ColumnDataSource for the plot
source = ColumnDataSource(data=dict(x=x, y=y))

# Create a new plot with a title and axis labels
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=700, height=400)

# Add a line renderer with legend and line thickness
p.line('x', 'y', source=source, legend_label="sin(x)", line_width=2)

# Show the plot in the notebook
output_notebook()
show(p)
```

### 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns



```python
from bokeh.transform import linear_cmap
from bokeh.palettes import Viridis256
import random

# Generate random data
n_points = 100
x_values = [random.uniform(0, 10) for _ in range(n_points)]
y_values = [random.uniform(0, 10) for _ in range(n_points)]
sizes = [random.uniform(5, 20) for _ in range(n_points)]  # random sizes for markers
colors = [random.uniform(0, 10) for _ in range(n_points)]  # values to map to colors

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x_values, y=y_values, sizes=sizes, colors=colors))

# Define color mapper based on the 'colors' column
color_mapper = linear_cmap(field_name='colors', palette=Viridis256, low=min(colors), high=max(colors))

# Create a scatter plot
scatter_plot = figure(title="Random Scatter Plot", x_axis_label='x', y_axis_label='y', width=700, height=400)

# Add scatter with different sizes and colors
scatter_plot.scatter('x', 'y', size='sizes', color=color_mapper, source=source, legend_label="Random Points", fill_alpha=0.6)

# Show the plot in the notebook
output_notebook()
show(scatter_plot)
```

### 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
# Prepare data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a ColumnDataSource for the plot
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create a figure for the bar chart
bar_chart = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruits', y_axis_label='Count', width=700, height=400)

# Add bars to the chart
bar_chart.vbar(x='fruits', top='counts', width=0.6, source=source, legend_label="Counts", color="navy", fill_alpha=0.7)

# Show the plot in the notebook
output_notebook()
show(bar_chart)
```

### 4. Create a Bokeh histogram to visualize the distribution of the given data

```python
# Generate random data for the histogram
data_hist = np.random.randn(1000)

# Create a histogram with 30 bins
hist, edges = np.histogram(data_hist, bins=30)

# Create a figure for the histogram
histogram = figure(title="Histogram", x_axis_label='Value', y_axis_label='Frequency', width=700, height=400)

# Add a quad glyph to represent the histogram bars
histogram.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], fill_color="blue", line_color="black", fill_alpha=0.7)

# Show the histogram plot in the notebook
output_notebook()
show(histogram)
```

### 5. Create a Bokeh heatmap using the provided dataset

```python
import numpy as np
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import LinearColorMapper, ColorBar
from bokeh.transform import transform
from bokeh.palettes import Viridis256

# Prepare the data
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create the figure
heatmap = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y', width=700, height=400)

# Color mapper
color_mapper = LinearColorMapper(palette=Viridis256, low=np.min(data_heatmap), high=np.max(data_heatmap))

# Add the image to the plot
heatmap.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
heatmap.add_layout(color_bar, 'right')

# Show the plot
output_notebook()
show(heatmap)
```
